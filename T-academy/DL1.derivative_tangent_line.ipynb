## 미분 구현

```python




def numerical_diff(f, x): # 미분 정의
    h = 1e-6 # 0.00001
    return (f(x+h) - f(x-h)) / (2*h)


def function_1(x):
    return 0.01*x**2 + 0.1*x 

def function_2(x):
  return 0.01*x**3 + 0.1*x**2 

```

## 접선 방정식 구하기

```python

def tangent_line(f, x1): 
    '''
    접선 방정식을 y = ax + b 로 표현
    기울기가 a이고 점(x1,y1)를 지나는 함수의 b값은
    y1 = a*x1 + b
    b = y1 - a*x1
    
    '''
    a = numerical_diff(f, x1)  # x지점에서 미분한 값이 그 지점의 접선 기울기이다
    y1 = f(x1)
    b = y1 - a*x1
    print('slope = %.2f' % a, 'y_bias = %.2f' % b)
    print('Y = %.2fX + %.2f'  % (a, b))
    print('Y = %fX + %f'  % (a, b))
    return lambda x: a*x + b
```

```python
numerical_diff(function_1, 5)
```

```python
numerical_diff(function_1, 10)
```

```python
numerical_diff(function_2, 10)
```

```python
x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)
plt.xlabel("x")
plt.ylabel("f(x)")

tf = tangent_line(function_1, 5)
y2 = tf(x)

plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```
